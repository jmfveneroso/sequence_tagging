### Questão 1

Teoria de Conjuntos: O objetivo é apenas verificar se você está informado sobre a diferença conceitual
entre conjuntos enumeráveis e não-enumeráveis. Não é necessário saber provar que um
conjunto é não-enumerável. Diga quais dos conjuntos abaixo é um conjunto enumerável e qual é
não-enumerável:

* $ \{0,1,2 \} $ = enumerável
* naturais:  $ \mathbb{N} = \{0,1,2,... \} $ = enumerável
* inteiros:  $ \mathbb{Z} = \{...,-2,-1,0,1,2,... \} $ = enumerável
* reticulado no plano inteiro: $ \{ (x,y); x \in \mathbb{Z}, y \in \mathbb{Z} \} $  = enumerável
* racionais: $ \mathbb{Q} = \{ p/q; q \gt 0, p \in \mathbb{Z}, q \in \mathbb{Z} \} $ = enumerável
* reais: $ \mathbb{R} $ = não enumerável
* irracionais: $ \mathbb{R} - \mathbb{Q} $ = não enumerável

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-1., 2., 0.01)

actual_fn = np.exp(x)
taylor_expansion_0 = 1 + x + x**2/2 
taylor_expansion_1 = np.exp(1)*(1 + (x-1) + (x-1)**2/2) 

plt.figure(figsize=(7, 7))
plt.plot(x, actual_fn, 'r')
plt.plot(x, taylor_expansion_0, 'g')
plt.plot(x, taylor_expansion_1, 'b')
plt.show()
```

### Questão 8

* a) O conjunto da combinação linear dos vetores $ v_1, ..., v_k $ se dá por 
$ C = \{ \alpha_1 v_1 + \alpha_2 v_2 + ... + \alpha_k v_k; \alpha_i \in \mathbb{R} \} $.

Considere dois vetores $ a $ e $ b $ tal que $ a, b \in C $, logo:

$ a = \alpha_1 v_1 + \alpha_2 v_2 + ... + \alpha_k v_k $ 

$ b = \beta_1 v_1 + \beta_2 v_2 + ... + \beta_k v_k $ 

Então, temos que:

$ c = a + b = (\alpha_1 + \beta_1)v_1 + (\alpha_2 + \beta_2)v_2 + ... + (\alpha_k + \beta_k)v_k $

Como $ \alpha_i + \beta_i \in \mathbb{R} \forall i $, então $ c \in C $

Considerando que:

$ d = ka; k \in \mathbb{R} $

$ d = k \alpha_1 v_1 + k \alpha_2 v_2 + ... + k \alpha_k v_k $

Como $ k \alpha_i \in \mathbb{R} \forall i $, então $ d \in C $
